```
stop = set(w for w in stopwords.words('english'))
```

```
cfd = nltk.ConditionalFreqDist((cat, w.lower())
                               for cat in movie_reviews.categories()
                               for w in movie_reviews.words(categories=cat)
                               if w.isalpha() and w.lower() not in stop)
```

```
cfd['pos'].most_common(10)
```

```
cfd['neg'].most_common(10)
```

```
diff = [w for w in cfd['pos'] if w not in cfd['neg']]
diff[:20]
```

```
from math import log2

N = cfd.N()
fd = cfd['pos'] + cfd['neg']
cat_N = {'pos':cfd['pos'].N(), 'neg':cfd['neg'].N()}

def pmi(word, cat):
  if cfd[cat][word] > 0:
    return log2((cfd[cat][word] * N)/(fd[word]*cat_N[cat]))
  else:
    return 0.0
```

```
mi = nltk.FreqDist()
for word in cfd['pos']:
  mi[word] = pmi(word, 'pos')
mi.most_common(10)  
```

```
mi = nltk.FreqDist()
for word in cfd['pos']:
  if cfd['pos'][word] > 25:
    mi[word] = pmi(word, 'pos')
mi.most_common(10) 
```

```
so = nltk.FreqDist()
for word in fd:
  if fd[word] > 25:  
    so[word] = pmi(word, 'pos') - pmi(word, 'neg')
so.most_common(20)
```

```
so.most_common()[-20:]
```
