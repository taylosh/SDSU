## Frequency dictionary

Count the number of tokens of each type 

```
fdist = nltk.FreqDist(words)
fdist.most_common(10)
```

```
fdist.plot(1000)
```

```
fdist.most_common(20)
```

For next time, we'll **normalize** the text: get rid of puncuation and make everything lower case

```
fdist = nltk.FreqDist([w.lower() for w in nltk.corpus.treebank if w.isalpha()])
```

```
def normalize(tokens):
  return [w.lower() for w in tokens if w.isalpha()]
```

```
fdist = nltk.FreqDist(normalize(nltk.corpus.brown.words()))
```

```
fdist.most_common(20)
```

```
fdist = nltk.FreqDist(normalize(nltk.corpus. brown.words()))
```
