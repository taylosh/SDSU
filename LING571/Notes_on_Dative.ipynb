```
import nltk
nltk.download('book', quiet=True)
```

```
from nltk.corpus import treebank
```

```
s = treebank.parsed_sents()[0]
s.pretty_print()
```

```
s2 = treebank.parsed_sents()[86]
s2.pretty_print()
```

```
from nltk import ParentedTree
from nltk.tgrep import tgrep_nodes
from toolz import concat
```

```
s2=ParentedTree.convert(s2)
```

```
for vp in concat(tgrep_nodes('VP', [s2])):
  vp.pretty_print()
```

```
for vp in concat(tgrep_nodes('VP <1 /^VB/ <2 NP <3 NP', [s2])):
  vp.pretty_print()
```

```
for vp in concat(tgrep_nodes('VP <1 /^VB/ <2 NP <3 NP', [s2])):
  print(vp.leaves())
  print(vp.leaves()[0])
```

```
sents = [ParentedTree.convert(s) for s in treebank.parsed_sents()]
```

```
fdative = nltk.FreqDist(vp.leaves()[0]
  for vp in concat(tgrep_nodes('VP <1 /^VB/ <2 NP <3 NP', sents)))
```

```
fdative.most_common()
```

```
wnl = nltk.WordNetLemmatizer()
```

```
wnl.lemmatize('gave', pos='v')
```

```
fdative2 = nltk.FreqDist(wnl.lemmatize(vp.leaves()[0].lower(), pos='v')
  for vp in concat(tgrep_nodes('VP <1 /^VB/ <2 NP <3 NP', sents)))
```

```
fdative2.most_common()
```
