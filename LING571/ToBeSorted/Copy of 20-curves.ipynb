```
import numpy as np

import matplotlib.pylab as plt
%config InlineBackend.figure_format = 'retina'
```

```
import nltk
nltk.download('book', quiet=True)
```

```
def normalize(words):
  return [w.lower() for w in words if w.isalpha()]
```

```
moby = normalize(nltk.corpus.gutenberg.words('melville-moby_dick.txt'))
paradise = normalize(nltk.corpus.gutenberg.words('milton-paradise.txt'))
stories = normalize(nltk.corpus.gutenberg.words('bryant-stories.txt'))
```

Python plotting functions take a list of *x* values and a list of *y* values, which is not very convenient for us.  So we can define a helper function that takes a list of *x,y* pairs and plots them.

```
def plotxy(pts, **kwargs):
  x, y = zip(*pts)
  plt.plot(x, y, **kwargs)
```

Plot curves for successive 5,000 word chunks of *Moby-Dick*

```
freq = nltk.FreqDist([len(w) for w in moby[:5000]])
plotxy(sorted(freq.items()))

freq = nltk.FreqDist([len(w) for w in moby[5000:10000]])
plotxy(sorted(freq.items()))

freq = nltk.FreqDist([len(w) for w in moby[10000:15000]])
plotxy(sorted(freq.items()))

freq = nltk.FreqDist([len(w) for w in moby[15000:20000]])
plotxy(sorted(freq.items()))

plt.title('Moby-Dick')
plt.xlabel('Length')
plt.ylabel('Count')

plt.show()
```

Plot curves for the first 20,000 words of *Moby-Dick*, *Paradise Lost*, and *Stories to Tell to Children*

```
freq = nltk.FreqDist([len(w) for w in moby[:20000]])
plotxy(sorted(freq.items()), label='Moby-Dick')

freq = nltk.FreqDist([len(w) for w in paradise[:20000]])
plotxy(sorted(freq.items()), label='Paradise Lost')

freq = nltk.FreqDist([len(w) for w in stories[:20000]])
plotxy(sorted(freq.items()), label='Stories to Tell to Children')

plt.title('Characteristic Curves of Composition')
plt.xlabel('Length')
plt.ylabel('Count')
plt.legend()

plt.show()
```
