```
orig = [t for t in tweets if 'retweeted_status' not in t]
```

```
len(orig)
```

```
orig[0]['text']
```

```
from nltk.tokenize.casual import TweetTokenizer
```

```
tokenizer = TweetTokenizer()
```

```
corpus = [[w.lower() for w in tokenizer.tokenize(t['text'])] for t in orig]
```

```
corpus[0]
```

```
corpus[6]
```

```
fd = nltk.FreqDist([w for t in corpus for w in t])
fd.most_common(10)
```

```
fd['cow'], fd['steer']
```

```
fsteer = nltk.FreqDist([w for t in corpus for w in t if 'steer' in t])
fsteer.most_common(10)
```

```
from math import log2
```

```
mi = nltk.FreqDist()
for w in fsteer:
  mi[w] = log2((fsteer[w]+fd.N())/(fd[w]+fd['steer']))
mi.most_common(5)
```

```
mi = nltk.FreqDist()
for w in fsteer:
  if fsteer[w] > 1:
    mi[w] = log2((fsteer[w]+fd.N())/(fd[w]+fd['steer']))
mi.most_common(10)
```

```
mi = nltk.FreqDist()
for w in fsteer:
  if fsteer[w] > 2:
    mi[w] = log2((fsteer[w]+fd.N())/(fd[w]+fd['steer']))
mi.most_common(10)
```

```
fbigram = nltk.FreqDist(b for t in corpus for b in nltk.bigrams(t))
fbigram.most_common(5)
```

```
ftrigram = nltk.FreqDist(b for t in corpus for b in nltk.trigrams(t))
ftrigram.most_common(5)
```
