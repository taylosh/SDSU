### Now let's group people by age.

```python
def agebreak(state):
    state['under20'] = state['age'].map(lambda x: x < 20)
    state['from20to50'] = state['age'].map(lambda x: 19 < x < 51)
    state['above50'] = state['age'].map(lambda x: x > 50)
```

```python
agebreak(p_df_AZ)
agebreak(p_df_CA)
agebreak(p_df_FL)
agebreak(p_df_HI)
agebreak(p_df_IL)
agebreak(p_df_KY)
agebreak(p_df_NV)
agebreak(p_df_NY)
agebreak(p_df_OR)
agebreak(p_df_TX)
```

### And we may as well map education and race while we are at it.

```python
def racemap(state):
    state['pacific_islander'] = state['race'].map(lambda x: '02' <= x <= '06')
    state['asian'] = state['race'].map(lambda x: '07' <= x <= '31')
    state['native_american'] = state['race'].map(lambda x: '32' <= x <= '36')
    state['black'] = state['race'].map(lambda x: '37' <= x <= '46')
    state['white'] = state['race'].map(lambda x: '47' <= x <= '70')
```

```python
racemap(p_df_AZ)
racemap(p_df_CA)
racemap(p_df_FL)
racemap(p_df_HI)
racemap(p_df_IL)
racemap(p_df_KY)
racemap(p_df_NV)
racemap(p_df_NY)
racemap(p_df_OR)
racemap(p_df_TX)
```

```python
p_df_HI[p_df_HI['pacific_islander']]
```

```python
def edmap(state):
    state['grade_school'] = state['education'].map(lambda x: '00' <= x <= '03')
    state['middle_school'] = state['education'].map(lambda x: x == '04')
    state['high_school_nd'] = state['education'].map(lambda x: '05' <= x <= '08')
    state['high_school_yd'] = state['education'].map(lambda x: x == '09')
    state['some_college'] = state['education'].map(lambda x: '10' <= x <= '11')
    state['AA'] = state['education'].map(lambda x: x == '12')
    state['BA'] = state['education'].map(lambda x: x == '13')
    state['MA'] = state['education'].map(lambda x: x == '14')
    state['prof_deg'] = state['education'].map(lambda x: x == '15')
    state['PhD'] = state['education'].map(lambda x: x == '16')
```

```python
edmap(p_df_AZ)
edmap(p_df_CA)
edmap(p_df_FL)
edmap(p_df_HI)
edmap(p_df_IL)
edmap(p_df_KY)
edmap(p_df_NV)
edmap(p_df_NY)
edmap(p_df_OR)
edmap(p_df_TX)
```

## Finally, we will look at income levels by education for all ten states

```python
def ed_lv (x):
    if '05' <= x <= '08': 
        return '1. High School-No Diploma'
    elif  x == '09':
        return '2. High School-Diploma'
    elif '10' <= x <= '11':
        return '3. Some College-No Degree'
    elif x == '12':
        return '4. AA Degree'
    elif  x == '13':
        return '5. BA Degree'
    elif x == '14':
        return '6. MA Degree'
    elif  x == '15':
        return '7. Professional Degree'
    elif x == '16':
        return '8. Phd Degree'
```

```python
AZ_inc = p_df_AZ[p_df_AZ['income']>0]
AZ_inc['ed_bylv'] = AZ_inc['education'].map(ed_lv)
AZ_incbyed = p_df_AZ['income'].groupby(AZ_inc['ed_bylv'])
AZ_incbyed_mean = AZ_incbyed.mean().round(2)

CA_inc = p_df_CA[p_df_CA['income']>0]
CA_inc['ed_bylv'] = CA_inc['education'].map(ed_lv)
CA_incbyed = p_df_CA['income'].groupby(CA_inc['ed_bylv'])
CA_incbyed_mean = CA_incbyed.mean().round(2)

FL_inc = p_df_FL[p_df_FL['income']>0]
FL_inc['ed_bylv'] = FL_inc['education'].map(ed_lv)
FL_incbyed = p_df_FL['income'].groupby(FL_inc['ed_bylv'])
FL_incbyed_mean = FL_incbyed.mean().round(2)

HI_inc = p_df_HI[p_df_HI['income']>0]
HI_inc['ed_bylv'] = HI_inc['education'].map(ed_lv)
HI_incbyed = p_df_HI['income'].groupby(HI_inc['ed_bylv'])
HI_incbyed_mean = HI_incbyed.mean().round(2)

IL_inc = p_df_IL[p_df_IL['income']>0]
IL_inc['ed_bylv'] = IL_inc['education'].map(ed_lv)
IL_incbyed = p_df_IL['income'].groupby(IL_inc['ed_bylv'])
IL_incbyed_mean = IL_incbyed.mean().round(2)

KY_inc = p_df_KY[p_df_KY['income']>0]
KY_inc['ed_bylv'] = KY_inc['education'].map(ed_lv)
KY_incbyed = p_df_KY['income'].groupby(KY_inc['ed_bylv'])
KY_incbyed_mean = KY_incbyed.mean().round(2)

NV_inc = p_df_NV[p_df_NV['income']>0]
NV_inc['ed_bylv'] = NV_inc['education'].map(ed_lv)
NV_incbyed = p_df_NV['income'].groupby(NV_inc['ed_bylv'])
NV_incbyed_mean = NV_incbyed.mean().round(2)

NY_inc = p_df_NY[p_df_NY['income']>0]
NY_inc['ed_bylv'] = NY_inc['education'].map(ed_lv)
NY_incbyed = p_df_NY['income'].groupby(NY_inc['ed_bylv'])
NY_incbyed_mean = NY_incbyed.mean().round(2)

OR_inc = p_df_OR[p_df_OR['income']>0]
OR_inc['ed_bylv'] = OR_inc['education'].map(ed_lv)
OR_incbyed = p_df_OR['income'].groupby(OR_inc['ed_bylv'])
OR_incbyed_mean = OR_incbyed.mean().round(2)

TX_inc = p_df_TX[p_df_TX['income']>0]
TX_inc['ed_bylv'] = TX_inc['education'].map(ed_lv)
TX_incbyed = p_df_TX['income'].groupby(TX_inc['ed_bylv'])
TX_incbyed_mean = TX_incbyed.mean().round(2)

```
